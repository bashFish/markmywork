```python
import docx
```

```python
def parse_file(filepath):
    doc = docx.Document(filepath)
    count = 0
    essay = ''
    count_paragraphs = 0
    for p in doc.paragraphs:
        if len(p.text) < 3:
            continue
        count += len(p.text)
        count_paragraphs += 1
        essay += p.text.strip().strip('\n')+'\n'
    return (filepath, count_paragraphs, count, essay)
parse_file(all_essays_filepaths[0])
```

```python
all_questions = []
for k in list(questions):
    all_questions.append(str(k))
all_questions
```

```python
all_parsed_essays = []
for fp in all_essays_filepaths:
    (filepath, count_paragraphs, count, essay) = parse_file(fp)
    #if count <= 3400: #its letter count, not token! darn.
    #    continue
    all_parsed_essays.append((fp, count, count_paragraphs, essay))
```

```python
all_answers = []
for i,essay_array in tqdm.tqdm(enumerate(all_parsed_essays)):
    essay_id = int(re.findall(r'\d+', essay_array[0])[-1])
    cur_row = filtered_marked_essays[filtered_marked_essays['essay id'] == essay_id]
    if len(cur_row) != 1:
        continue

    for j,k in enumerate(all_questions):
        
        prompt = 'Given is a student essay and a question to the essay. Please rate the Question with a number between 0 and 3, where 0 is "not at all satisfied" and 3 is "very satisfied.".\n\n'
        prompt += f'Essay: """\n{essay_array[3][:-1]}"""\n' #implicit newline
        prompt += f'Question: """\n{str(k).strip()}"""\n'
        prompt += 'Answer:'
        #print(prompt)

        #response = openai.Completion.create(model="text-davinci-003", prompt=prompt, temperature=0, max_tokens=3)
        all_answers.append((i, j, prompt, response, int(cur_row[k])))
        #time.sleep(.5)
        #print(f"{i} {response['choices'][0]['text']}")
```
