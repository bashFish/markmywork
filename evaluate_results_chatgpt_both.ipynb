```python
import re
```

```python
def parse_float(text):
    pattern = "[+-]?([0-9]+([.][0-9]*)?|[.][0-9]+)"
    match = re.search(pattern, text)
    if match:
        float_str = match.group()
        try:
            return int(float_str) if '.' not in float_str else float(float_str)
        except ValueError:
            return None
    else:
        return None
```

```python
ratings = []
for a in all_answers:
    answer = a[3]["choices"][0]["message"]["content"]
    target_grade = a[-1]
    
    if "Grade:" in answer:
        grade = answer.split("Grade:")[1]
        grade = float(grade[:4].strip().split('/')[0].split('(')[0])
        
        ratings.append((grade, target_grade, a[1], a[0]))
    elif answer.count('out of') == 1:
        grade = answer.split('out of')[0][-4:].strip()
        grade = parse_float(grade)
        
        if grade is None:
            continue
        ratings.append((grade, target_grade, a[1], a[0]))
```
